```python
df_final=df.drop(['PassengerId','Pclass',"Name",'Ticket','Embarked','Sex','SibSp','Parch'],axis=1)
df_test=df_test.drop(['PassengerId','Pclass',"Name",'Ticket','Embarked','Sex','SibSp','Parch'],axis=1)
```

```python
train=df_final.iloc[:,1:]
```

```python
test=df_final.iloc[:,0]
```

```python
test.head()
```

```python
train['Age']=(train['Age'])/train['Age'].std()
train['Fare']=(train['Fare'])/train['Fare'].std()
df_test['Age']=(df_test['Age'])/df_test['Age'].std()
df_test['Fare']=(df_test['Fare'])/df_test['Fare'].std()
test=np.array(test)
train=np.array(train)
```

```python
from keras.models import Sequential
```

```python
model=Sequential([
                
                keras.layers.Dense(32,activation='relu',input_dim=20),
                keras.layers.Dense(64,activation='relu'),
                keras.layers.Dense(128,activation='relu'),
                keras.layers.Dense(256,activation='relu'),
                keras.layers.Dense(1,activation='sigmoid')
])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.fit(train,test,epochs=100)
```
